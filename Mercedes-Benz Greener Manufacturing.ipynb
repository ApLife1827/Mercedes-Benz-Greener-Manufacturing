```python
#Drop Id
train=train.drop(['ID'], axis = 1) 
id_test=test['ID']
test=test.drop(['ID'], axis = 1)
```

```python
train=train.loc[:, (train != train.iloc[0]).any()]
test=test.loc[:, (test != test.iloc[0]).any()]
```

```python
y_train=train['y']
col_train=train.columns
col_test=test.columns
col=[]
for i in col_train:
    for j in col_test:
        if(i==j):
            col.append(i)
```

```python
#Checking missing values
train=train[col]
test=test[col]
train.isnull().sum()/len(train)*100
```

```python
test.isnull().sum()/len(test)*100
```

```python
# label encoder
train.head()
```

```python
obj_col=['X0','X1','X2','X3','X4','X5','X6','X8']
```

```python
#Label Encoding for object to numeric conversion
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in obj_col:
    train[feat] = le.fit_transform(train[feat].astype(str))

print (train.info())
```

```python
train.head()
```

```python
for feat in obj_col:
    test[feat] = le.fit_transform(test[feat].astype(str))

print (test.info())
```

```python
test.head()
```

```python
# performing preprocessing part
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(train)
X_test = sc.fit_transform(test)
```

```python
# Applying PCA function on training
# and testing set of X component
from sklearn.decomposition import PCA

pca = PCA(n_components =12)

X_train = pca.fit_transform(X_train)
X_tests = pca.transform(X_test)

explained_variance = pca.explained_variance_ratio_
```

```python
import xgboost as xgb
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
        X_train, 
        y_train, test_size=0.3)
```

```python
xg_train = xgb.DMatrix(X_train, label=y_train)
xg_test = xgb.DMatrix(X_test, label=y_test)
xg_tests = xgb.DMatrix(X_tests)
```

```python
params = {
    'colsample_bytree': 0.8,
    'objective': 'reg:linear',
    'max_depth': 7,
    'min_child_weight': 1,
    'learning_rate': 0.1,
    'subsample': 0.8,
    'num_class':1,
    'eta': 0.2
}

model = xgb.cv(
    params=params,
    dtrain=xg_train,
    num_boost_round=500,
    nfold=5,
    early_stopping_rounds=100
)
# Fit
final_gb = xgb.train(params, xg_train, num_boost_round=len(model))
```

```python
p_test = final_gb.predict(xg_tests)

sub = pd.DataFrame()
sub['ID'] = id_test
sub['y'] = p_test
sub.to_csv('xgb.csv', index=False)

sub.head()
```
